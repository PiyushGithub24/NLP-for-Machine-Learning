# Named Entity Recognition :

Named Entity Recognition (NER) is an information extraction technique in Natural Language Processing (NLP) that identifies and classifies named entities in text into predefined categories such as:

```python
sentence="Microsoft was founded by Bill Gates and Paul Allen in 1975 in Albuquerque,\
New Mexico. The company later moved its headquarters to Redmond, Washington. In 2023,\
Microsoft reported a revenue of $211 billion."
"""
NER Output:
Microsoft → [ORG] (Organization)
Bill Gates → [PERSON] (Person)
Paul Allen → [PERSON] (Person)
1975 → [DATE] (Date)
Albuquerque, New Mexico → [GPE] (Location)
Redmond, Washington → [GPE] (Location)
2023 → [DATE] (Date)
$211 billion → [MONEY] (Monetary Value)"""
```

```python
sentence
```

```python
import nltk
words=nltk.word_tokenize(sentence)
```

```python
words
```

```python
tag_elements=nltk.pos_tag(words)
```

```python
nltk.ne_chunk(tag_elements).draw()
```
